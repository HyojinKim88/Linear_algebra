# 역행렬
$AA^{-1} = A^{-1}A = I$
* det(A)=0 이면 역행렬은 존재하지 않음
<=> A는 특이 행렬(singular matrix)
* $A^{-1}$가 존재하는 경우 : A는 가역 행렬(invertible matrix), 가역 행렬의 역행렬은 유일함

## 역행렬의 계산
IMG_BE41B4BB32A3-1.jpeg

## 정방 행렬A의 거듭제곱
- $A^0 = I$
- $A^n = AA ∙∙∙ A$
- $A^nA^m = A^{n+m}$, $(A^n)^m = A^{nm}$

## 역행렬의 성질 (A는 가역 행렬)
- $A^{-n} = (A^{-1})^n = A^{-1}A^{-1}∙∙∙A^{-1}$
- $(A^{-1})^{-1} = A$
- $(A^n)^{-1} = (A^{-1})^n$
- $(aA)^{-1} = \frac{1}{a}A^{-1} $
----------
-  $(A^{-1})^T = (A^T)^{-1}$
------
- $AA^T$과 $A^TA$도 가역행렬!
-------
- $det(A^{-1})$ = $\frac{1}{det(A)}$
  - $AA^{-1}$=$A^{-1}A$=$I$

```python
import numpy as np

A = np.array([[3,2,0],
              [-1,-3,6],
              [2,3,-5]])
invA = np.linalg.inv(A)
print(A)
print("-----")
print(invA)
```
