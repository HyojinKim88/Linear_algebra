# 행렬식의 성질

* 삼각 행렬의 행렬식 : $det(A)$ = 대각원소들의 곱
* 대각 행렬의 행렬식 : $det(A)$ = 대각원소들의 곱
* 전치 행렬의 행렬식 : $det(A)$ = $det(A^T)$
* 특정 행 또는 열의 원소가 모두 0일때 : $det(A)$=0

## 행렬의 기본행 연산과 행렬식 
1. 한 행에 0이 아닌 상수를 모두 곱한다
* A의 특정 행 또는 열을 k배 한 행렬을 B라고 할 때,<br>
$k * det(A) = det(B)$
* A의 모든 원소에 k배를 하면
$det(kA) = k^n det(A)$
2. 두 행을 교환한다
* A의 두 행을 교환한 행렬을 B라고 할 때
$-det(A) = det(B)$
3. 한 행의 배수를 다른 행에 더한다.
* A의 한 행의 배수를 다른 행에 더한 행렬을 B라고 할 때
$det(A) = det(B)$
-------------------------------
* A의 한 행 또는 열이 다른 행 또는 열의 배수일 때<br>
$det(A) = 0$
---------------------
* 행렬곱과 행렬식<br>
$det(AB) = det(A)det(B)$

```python
import numpy as np

A = np.array([[3,2,0],
              [-1,-3,6],
              [2,3,-5]])
detA = np.linalg.det(A)
print(detA)
#전치행렬
detAT = np.linalg.det(A.T)
print(detAT)
```

```python
# 삼각행렬
A = np.array([[3,2,0],
              [-1,-3,6],
              [2,3,-5]])
At = np.tril(A)
detAt = np.linalg.det(At)
print(detAt) # 부동 소수점 연산
```

```python
# 특정행이 0
A = np.array([[0,0,0],
              [-1,-3,6],
              [2,3,-5]])
detA = np.linalg.det(A)
print(detA)
```

```python
#행렬의 기본행 연산과 행렬식
A = np.array([[3,2,0],
              [-1,-3,6],
              [2,3,-5]])
detA = np.linalg.det(A)
print(detA)
print("==============")
print("1-1. A의 특정 행 또는 열을 k배 한 행렬을 B라고 할 때, 𝑘∗𝑑𝑒𝑡(𝐴)=𝑑𝑒𝑡(𝐵)") 
B = np.array([[6,4,0], #1행에 2배
              [-1,-3,6],
              [2,3,-5]])
detB = np.linalg.det(B)
print(detB)
print("==============")
print("1-2. A의 모든 원소에 k배를 하면 𝑑𝑒𝑡(𝑘𝐴)=𝑘^𝑛*𝑑𝑒𝑡(𝐴)") 
det2A = np.linalg.det(2*A)
print(det2A)
print(2*2*2*detA)
print("==============")
print("2. A의 두 행을 교환한 행렬을 B라고 할 때  −𝑑𝑒𝑡(𝐴)=𝑑𝑒𝑡(𝐵)") 
changeB = np.array([[-1,-3,6], #1,2행 교환
              [3,2,0],
              [2,3,-5]])
detcB = np.linalg.det(changeB)
print(detcB)
print("==============")
print("3. A의 한 행의 배수를 다른 행에 더한 행렬을 B라고 할 때  𝑑𝑒𝑡(𝐴)=𝑑𝑒𝑡(𝐵)") 
newB = np.array([[3,2,0], 
              [5,1,6], #1행에 2배를 2행에 더함
              [2,3,-5]])
detnB = np.linalg.det(newB)
print(detnB)
```

```python
A = np.array([[3,2,0],
              [-1,-3,6],
              [2,3,-5]])
detA = np.linalg.det(A)
print(detA)
print("A의 한 행 또는 열이 다른 행 또는 열의 배수일 때 𝑑𝑒𝑡(𝐴)=0 ")
AA = np.array([[3,2,0],
              [6,4,0], # 1행의 배수
              [2,3,-5]])
detAA = np.linalg.det(AA)
print(AA)
print("====================")
print("𝑑𝑒𝑡(𝐴𝐵)=𝑑𝑒𝑡(𝐴)𝑑𝑒𝑡(𝐵) ")
B = np.array([[3,1,2],
              [1,0,-1],
              [2,0,-1]])
detB = np.linalg.det(B)
print(detB)
AB = np.matmul(A,B)
print(AB)
detAB = np.linalg.det(AB)
print(f"detA * detB = {detAB}")
```
